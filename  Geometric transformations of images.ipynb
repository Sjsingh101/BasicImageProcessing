#  Geometric transformations of images - scaling, translation, rotation, perspective transformation.

**Scaling :**

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv.imread("landscape.jpg")
```

```python
plt.axis("off")
plt.imshow(image)
```

```python
row,column,_ = image.shape
print("Rows : {} \nColumns : {}".format(row,column))
```

```python
scaled_image = cv.resize(image,None,fx=0.5,fy=1.5)

plt.axis("off")
plt.imshow(scaled_image)
```

**Traslation :**

logic of matrix

```python
matt = np.array([[1, 0, 50], [0, 1, 50]],dtype=np.float64)
```

```python
trans_image = cv.warpAffine(image, matt, (column, row))

plt.axis("off")
plt.imshow(trans_image)
```

**Rotation :**

```python
matr = cv.getRotationMatrix2D((column/2, row/2), 60, 1)
```

```python
rotated_image = cv.warpAffine(image, matr, (column, row))

plt.axis("off")
plt.imshow(rotated_image)
```
