#  Dilation, Erosion, Closing and Opening based image morphology

## Dilation 

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread("orange.jpg")

plt.axis("off")
plt.imshow(img)
```

```python
_,mask=cv2.threshold(img,230,255,cv2.THRESH_BINARY_INV)

plt.axis("off")
plt.imshow(mask)
```

```python
kernel=np.ones((5,5),dtype=np.float64)
dilated_img=cv2.dilate(mask,kernel)

plt.axis("off")
plt.imshow(dilated_img)
```

## Erosion

```python
erosion_img=cv2.erode(mask,kernel)

plt.axis("off")
plt.imshow(erosion_img)
```

## Opening

```python
opening = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

plt.axis("off")
plt.imshow(opening)
```

## Closing

```python
closing = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

plt.axis("off")
plt.imshow(closing)
```
