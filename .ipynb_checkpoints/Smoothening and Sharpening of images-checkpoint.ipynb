# Smoothening and Sharpening of images

## Smoothening of images

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv.imread("noisy1.png")

plt.axis("off")
plt.imshow(image)
```

```python
image1 = cv.imread("noisy.jpg")

plt.axis("off")
plt.imshow(image1)
```

```python
blur_average = cv.blur(image, (5, 5))

plt.axis("off")
plt.imshow(blur_average)
```

```python
blur_gaussian = cv.GaussianBlur(image, (5, 5), 0)

plt.axis("off")
plt.imshow(blur_gaussian)
```

```python
blur_gaussian1 = cv.GaussianBlur(image1, (5, 5), 0)

plt.axis("off")
plt.imshow(blur_gaussian1)
```

```python
blur_median = cv.medianBlur(image, 5)

plt.axis("off")
plt.imshow(blur_median)
```

```python
blur_bilateral = cv.bilateralFilter(image, 5, 45, 45)

plt.axis("off")
plt.imshow(blur_bilateral)
```

## Sharpening of images

```python
sum = cv.addWeighted(image1,1.5,blur_gaussian1,-0.5,0)

plt.axis("off")
plt.imshow(sum)
```
