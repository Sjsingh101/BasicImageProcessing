# Implement the functioning of a mask which convolves over an image

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv.imread("dog.jpg")
image.shape
```

```python
plt.axis("off")
plt.imshow(image)
```

```python
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
plt.axis("off")
plt.imshow(gray_image,cmap="gray")
```

```python
_, mask = cv.threshold(gray_image, 175, 255, cv.THRESH_BINARY)
```

```python
plt.axis("off")
plt.imshow(mask,cmap="gray")
```

```python
sum = cv.add(gray_image, gray_image, mask=mask)
```

```python
plt.axis("off")
plt.imshow(sum,cmap="gray")
```
