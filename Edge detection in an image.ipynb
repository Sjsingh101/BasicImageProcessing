# Edge detection in an image

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv.imread("landscape.jpg",cv.IMREAD_GRAYSCALE)
```

```python
plt.axis("off")
plt.imshow(image,cmap="gray")
```

```python
gaussian_blur = cv.GaussianBlur(image, (5, 5), 0)

plt.axis("off")
plt.imshow(gaussian_blur,cmap="gray")
```

```python
sobelx_edge = cv.Sobel(gaussian_blur, cv.CV_64F, 1, 0)

plt.axis("off")
plt.imshow(sobelx_edge,cmap="gray")
```

```python
sobely_edge = cv.Sobel(gaussian_blur, cv.CV_64F, 0, 1)

plt.axis("off")
plt.imshow(sobely,cmap="gray")
```

```python
laplacian_edge = cv.Laplacian(gaussian_blur, cv.CV_64F, ksize=5)

plt.axis("off")
plt.imshow(laplacian_edge,cmap="gray")
```

```python
canny_edge = cv.Canny(gaussian_blur, 123, 164)

plt.axis("off")
plt.imshow(canny_edge,cmap="gray")
```
